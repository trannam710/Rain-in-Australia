```python
for location in set(df['Location']):
  locate = []
  for i in range(0, len(df)):
    if df.loc[i,'Location'] == location:
      locate.append(df.iloc[i,:])
  locate = pd.DataFrame(list(locate), columns=df.columns)
  locate.to_csv('{}.csv'.format(location))
```

```python
# Hàm thêm biến Day để xét tính missing theo các tháng cho bước tiếp theo

def missing_follow_month(df):
  df['Day'] = ''
  df['Month'] = ''
  df['Year'] = ''
  for i in df.index:
    dtime = datetime.datetime.strptime(df['Date'][i], "%Y-%m-%d")
    df['Day'][i] = dtime.day
    df['Month'][i] = dtime.month
    df['Year'][i] = dtime.year

  return df
  
```

```python
# Lưu lại các location

for i in range(len(df)):
  df[i] = missing_follow_month(df[i])
  df[i].drop(columns=['Sunshine', 'Evaporation', 'Cloud3pm', 'Cloud9am'], axis=1, inplace=True)
  df[i].to_csv('/content/drive/MyDrive/Source code/Dataset/Dataset Giai Đoạn 1/{}.csv'.format(location[i]))
```
